## Introduction

This tutorial introduces FLAML (Fast and Lightweight AutoML), a powerful and efficient AutoML library developed by Microsoft. FLAML is designed to find accurate machine learning models within computational constraints, making it ideal for resource-limited environments.

In this tutorial, we will:

1. Understand how FLAML works compared to other AutoML tools
2. Apply FLAML to a materials science dataset on polymer crystallization tendency
3. Train multiple ML models automatically with minimal configuration
4. Evaluate and visualize model performance

## Key Features of FLAML

- **Efficiency**: Optimizes for both model accuracy and computational efficiency
- **Flexibility**: Supports various ML tasks (classification, regression, etc.)
- **Customizability**: Easily extensible to add custom learners or metrics
- **Cost-Effectiveness**: Uses techniques like cost-based search for faster convergence

## AutoML with FLAML

FLAML (Fast and Lightweight AutoML) is an efficient AutoML library that:
1. Automatically selects machine learning algorithms
2. Optimizes hyperparameters efficiently
3. Manages computational resources effectively
4. Provides simple APIs for easy use
